```python
import requests
from bs4 import BeautifulSoup as bs
import re
import time, random
import pandas as pd
```

```python
## Unpickle the data
df_21 = pd.read_pickle('2021 movie data part 1')
df_21.head()
```

```python
## Unpickle the data
df_21_2 = pd.read_pickle('2021 movie data part 2')
df_21_2.tail()
```

```python
## Concatenate data
all_2021 = pd.concat([df_21,df_21_2], axis=0, ignore_index=True)
all_2021.tail()
```

```python
## Unpickle the data 2020
df_20 = pd.read_pickle('2020 movie data')
df_20.tail(10)
```

```python
## Concatenate data
## It worked! Columns lined up!
all_20_21 = pd.concat([all_2021,df_20], axis=0, ignore_index=True)
all_20_21.tail(10)
```

```python
## Unpickle the data 2019
df_19 = pd.read_pickle('2019 movie data')
df_19.tail(5)
```

```python
## Concatenate data
all_21_20_19 = pd.concat([all_20_21,df_19], axis=0, ignore_index=True)
all_21_20_19.tail()
```

```python
## Unpickle the data 2018
df_18 = pd.read_pickle('2018 movie data')
df_18.tail(5)
```

```python
## Concatenate data
all_21_20_19_18 = pd.concat([all_21_20_19,df_18], axis=0, ignore_index=True)
all_21_20_19_18.tail()
```

```python
## Unpickle the data 2017
df_17 = pd.read_pickle('2017 movie data')
df_17.tail(5)
```

```python
## Concatenate data
all_21_20_19_18_17 = pd.concat([all_21_20_19_18,df_17], axis=0, ignore_index=True)
all_21_20_19_18_17.tail()
```

```python
## Unpickle the data 2016
df_16 = pd.read_pickle('2016 movie data')
df_16.tail(5)
```

```python
## Concatenate data
all_data = pd.concat([all_21_20_19_18_17,df_16], axis=0, ignore_index=True)
all_data.tail()
```

```python
## Pickle this dataframe!
all_data.to_pickle('movie_data_combined_years')
```

```python
print(all_data['Budget'].isna().sum())
print(all_data.shape)
```

```python
## Cleaning
all_data['domestic_dollars'] = all_data.Domestic.str.replace('$', '')
all_data['international_dollars'] = all_data.International.str.replace('$', '')
all_data['opening_weekend_dollars'] = all_data.International.str.replace('$', '')
all_data['released_theatres'] = all_data['Widest Release'].str.replace('theaters', '')
all_data['days_in_release'] = all_data['In Release'].str.replace('days', '')

all_data.head()


```

```python
all_data.shape
```

```python
## Drop movies not realeased inernationally
all_data.drop(all_data[all_data['Worldwide'] == 'Null'].index, inplace = True)
all_data.shape
```

```python
num_variables = all_data[['Name', 'domestic_dollars', 'international_dollars', 'opening_weekend_dollars', 'released_theatres', 'days_in_release']]

num_variables.loc[:, 'domestic_dollars'] = num_variables['domestic_dollars'].apply(lambda x: x.replace('$', ''))
num_variables.loc[:, 'international_dollars'] = num_variables.international_dollars.str.replace(',', '')
num_variables.loc[:, 'opening_weekend_dollars'] = num_variables.opening_weekend_dollars.str.replace(',', '')
num_variables.loc[:,'released_theatres'] = num_variables.released_theatres.str.replace(',', '')
num_variables.loc[:,'days_in_release'] = num_variables.days_in_release.str.replace(',', '')


num_variables.dropna(inplace = True)
num_variables.info()
```

```python
num_variables = num_variables.astype({'domestic_dollars': 'int', 'international_dollars': 'int', 
                                     'opening_weekend_dollars': 'int', 'released_theatres':'int', 'days_in_release': 'int'})
num_variables.info()
```

```python
num_variables.to_pickle('MVP_data')
```
